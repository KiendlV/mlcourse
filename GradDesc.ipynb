```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## gradient descend algo ##
```

```python
x = np.arange(0,10)
y = np.arange(0,10)
```

```python
plt.scatter(x,y)
```

```python
params = {"w" : -3,
          "b" : 5}
```

```python
def func(params,x):
    y = params["w"]*x+params["b"]
    return y
```

```python
def msee(y_true,y_pred):
    error = np.sum((y_true - y_pred)**2)
    return error/len(y)
```

```python
def deriv(y_true,y_pred):
    dw = np.sum(2*(y_true-y_pred)*-x)/len(y_true)

    db = np.sum(2*(y_true-y_pred))/len(y_true)
    return dw, db
```

```python
np.sum((y-func(params,x))**2)/len(y)
```

```python
np.sum(2*(y-func(params,x))*-x)
```

```python
def update(dw,db,lr=0.01):
    params["w"] = params["w"] - lr*dw
    params["b"] = params["b"] - lr*db
```

```python
loss = []
i = 0 
for i in range(10):
    if i < 100:
        y_pred = func(params,x)
        mse = msee(y,y_pred)
        loss.append(mse)
        dw, db = deriv(y,y_pred)
        update(dw,db)
        print(params)
        i = i + 1
```

```python
plt.plot(loss)
```
